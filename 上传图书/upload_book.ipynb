```python
import requests
import random

# 使用如下函数即可在一个列表中随机取出一个
users = ['zyx', 'yzm', 'stm']
stores = {
    'zyx': ['648宇宙连锁店', '648浙江分店', '648台州分店', 'ECNU书店', '终点书店'],
    'yzm': ['648全球连锁店', '648大夏书屋', '648哲哥书屋', '名学长有书卖', '客官请慢走'],
    'stm': ['648东北分店', '648吉林分店', '震惊华师第一店', '斗宗强者恐怖如斯', '数理恶魔拉格朗日书店'],
}

header = {
    'Content-Type': 'application/json',
}

for book_json in books_dic:
    user = random.choice(users)
    store = random.choice(stores[user])
    data = {
        'user_id': user,        # <- 可以在storeAPP中把判断userid的地方注释掉，现已注释token和checklogin
        'store_id': store,      # <- 然后就可以取出数据库所有店铺然后随机取一个，对其添加书本
        'book_info': book_json,
        'stock_level': random.randint(1, 648),
    }

    print(data['user_id'])
    print(data['store_id'])

    print(book_json['title'])
    print(book_json['original_title'])

                                    # 本地测试时，ip修改为本地的
    response = requests.post('http://139.196.42.152//seller/add_book/', json=(data), headers=header)

    print(response) # response会输出错误码
    print()
    
    # 注释这行开始循环
    # break
```
